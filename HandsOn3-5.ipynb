```python
from pyspark.sql import SparkSession as SS, functions as F
from pyspark.sql.window import Window

spark = SS.builder.appName("DatasetGermanCreditRisk").getOrCreate()
windowSpec = Window.partitionBy("Housing").orderBy("Credit amount")

df = spark.read.csv("german_credit_data.csv", header=True, inferSchema=True)
df.show()

# Menampilkan data jenis kelamin, umur, dan pekerjaan
df.select("Sex","Age","Job").show()

# Menampilkan data yang memiliki umur di atas 35
df.filter(df["Age"] > 35).show()

# Menjumlahkan total kredit berdasarkan jenis kelamin
df.groupBy("Sex").sum("Credit amount").show()

# Menambahkan kolom Born yang berisi tahun kelahiran
df.withColumn("Born", 2025 - df["Age"]).show()

# Mengurutkan data berdasarkan kategori Housing dan mengurutkannya lagi berdasarkan Credit amount
df.withColumn(("Rank"), F.rank().over(windowSpec)).show()
```
