```python
import pandas as pd
import matplotlib.pyplot as plt

# Mengubah SQL ke Pandas
gaji = spark.sql('''
SELECT d.DeptName, SUM(e.Salary) AS Total_Gaji
FROM employees e
JOIN departments d ON e.DeptId = d.DeptId
GROUP BY d.DeptName
ORDER BY d.DeptName
''').toPandas()

# Visualisasi
plt.figure(figsize=(8,8))
plt.pie(gaji['Total_Gaji'], labels=gaji['DeptName'], autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Persentase Gaji per Departemen')
plt.axis('equal')  
plt.show()
```
