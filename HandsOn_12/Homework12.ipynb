```python
import pandas as pd

# Load data
data = pd.read_csv('mening missing 12.csv')
df = pd.DataFrame(data)

# Identify missing values
missing_values = df.isnull()
print(missing_values.any())

# Show missing values
df[missing_values.any(axis=1)]
```

```python
# Filling missing values with mean and mode
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['WBC_Count'].fillna(df['WBC_Count'].mean(), inplace=True)
df['Protein_Level'].fillna(df['Protein_Level'].mean(), inplace=True)
df['Glucose_Level'].fillna(df['Glucose_Level'].mean(), inplace=True)
df['Pathogen_Present'].fillna(df['Pathogen_Present'].mode()[0], inplace=True)
df['Diagnosis'].fillna(df['Diagnosis'].mode()[0], inplace=True)
df['Outcome'].fillna(df['Outcome'].mode()[0], inplace=True)
df['Hemoglobin'].fillna(df['Hemoglobin'].mean(), inplace=True)
df['WBC_Blood_Count'].fillna(df['WBC_Blood_Count'].mean(), inplace=True)
df['Platelets'].fillna(df['Platelets'].mean(), inplace=True)
df['CRP_Level'].fillna(df['CRP_Level'].mean(), inplace=True)
```

```python
# Transformation to normalize numerical columns
df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 0}).astype(int)
df['Pathogen_Present'] = df['Pathogen_Present'].replace({'No': 0, 'Yes': 1}).astype(int)
df['Diagnosis'] = df['Diagnosis'].replace({'Unknown': 0, 'Viral': 1, 'Bacterial':2}).astype(int)
df['Outcome'] = df['Outcome'].replace({'Recovered': 1, 'Deceased': 0}).astype(int)
df['Risk_Level'] = df['Risk_Level'].replace({'Low Risk': 0, 'Moderate Risk': 1, 'High Risk':2}).astype(int)
```

```python
import numpy as np

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter untuk menghapus outlier
df_no_outlier_age = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

print("Jumlah data awal:", len(df))
print("Jumlah data setelah menghapus outlier Age:", len(df_no_outlier_age))
```
