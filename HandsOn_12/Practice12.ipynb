```python
import pandas as pd

# Load csv
data = pd.read_csv('Titanic-Dataset.csv')
df = pd.DataFrame(data)

# Identify missing values
missing_values = df.isnull()
print(missing_values.any())

# Show missing values
df[missing_values.any(axis=1)]
```

```python
# Filling missing values with mean and mode
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column
df = df.drop(columns=['Cabin'], errors='ignore')
```

```python
# Transformation to normalize numerical columns
df['Sex'] = df['Sex'].replace({'male': 1, 'female': 0}).astype(int)
df['Embarked'] = df['Embarked'].replace({'S': 0, 'C': 1, 'Q':2}).astype(int)
```

```python
# Check duplicate data
df[df.duplicated()]
```
