```python
from collections import defaultdict

def map_function(text):
    for word in text.split():
        yield (word, 1)

def reduce_function(pairs):
    result = defaultdict(int)
    for word, count in pairs:
        result[word] += count
    return dict(result)

teks = (
    "Di Untidar, mahasiswa belajar, belajar bersama, dan belajar setiap hari untuk berkembang. "
    "Di Untidar, setiap pengalaman, setiap kegiatan, dan setiap teman memberi pelajaran berharga. "
    "Di Untidar, dosen membimbing, kampus mendukung, dan fasilitas membantu mahasiswa belajar dengan lebih baik. "
    "Di Untidar, kegiatan akademik dan non-akademik berjalan bersamaan; mahasiswa berkreasi, berinovasi, "
    "dan berproses menjadi pribadi yang lebih baik. Di Untidar, persahabatan tumbuh, kerja sama terjalin, "
    "dan semangat belajar selalu hadir. Di Untidar, setiap langkah di kampus adalah kesempatan, "
    "setiap kesempatan di kampus adalah pembelajaran, dan setiap pembelajaran di kampus Untidar membentuk masa depan mahasiswa. "
    "Di Untidar, mahasiswa tidak hanya menuntut ilmu, tetapi juga membangun karakter, membangun pengalaman, "
    "dan membangun kenangan yang tak terlupakan."
)

mapF = map_function(teks) 
reduceF = reduce_function(mapF)
for word, count in sorted(reduceF.items()):
    print(f"{word}: {count}")
```
