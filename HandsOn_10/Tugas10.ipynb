```python
from prettytable import PrettyTable

# Menampilkan 5 karyawan dengan gaji tertinggi per departemen
pipeline = [
    {'$sort': {'department': 1, 'salary': -1}},  # Urutkan berdasarkan departemen & gaji tertinggi
    {'$group': {
        '_id': '$department',
        'top_employees': {'$push': {'name': '$name', 'salary': '$salary'}}
    }},
    {'$project': {'top_5': {'$slice': ['$top_employees', 5]}}}  # Ambil hanya 5 teratas
]

# Menampilkan hasil dalam tabel
for result in employees.aggregate(pipeline):
    dept = result['_id']
    table = PrettyTable()
    table.field_names = ["Nama", "Gaji"]
    for emp in result['top_5']:
        table.add_row([emp['name'], emp['salary']])
    print(f"\n=== Departemen: {dept} ===")
    print(table)
```

```python
# Menghapus semua karyawan yang berusia di bawah 25 tahun
delete_result = employees.delete_many({'age': {'$lt': 25}})

print(f"{delete_result.deleted_count} data karyawan di bawah 25 tahun telah dihapus.")
```

```python
# Laporan total gaji dan rata-rata umur per departemen
pipeline = [
    {'$group': {
        '_id': '$department',
        'total_salary': {'$sum': '$salary'},
        'average_age': {'$avg': '$age'}
    }}
]

for result in employees.aggregate(pipeline):
    print(f"Departemen: {result['_id']}")
    print(f"Total Gaji: {result['total_salary']}")
    print(f"Rata-rata Umur: {round(result['average_age'], 2)}\n")
```
